# Matplotlib是python中最常用的一个画图包，通常配合Numpy一起使用，可视化工具，常用于数据分析，非常方便

官网document: https://matplotlib.org/tutorials/index.html

eg:散点图Scatter分析两个变量之间的关系

```python
#导入第三方包
import numpy as np
import matplotlib.pyplot as plt

#这一句是因为要在jupyter notebook里操作，才要加上这句，否则不会显示出图
%matplotlib inline

#简单构造数据，x为1，2，3，4，5，6，7，8; y=x
x = np.arange(1,10)
y1 = x
y2 = 1.5*x


fig = plt.figure()
ax1 = fig.add_subplot(111)
#设置标题
ax1.set_title('Scatter Plot')
#设置X轴标签
plt.xlabel('Xtest')
#设置Y轴标签
plt.ylabel('Ytest')
#画散点图
lValue = x
#x,y分别代表横纵轴，c标识颜色（r即红色），s, linewidth标识了marker的宽度，这里是x值越大，marker越大。marker为散点的形状
ax1.scatter(x,y1,c='r',linewidths=lValue,marker='o')
ax1.scatter(x,y2,c='b',linewidths=lValue,marker='v')
#设置图标
plt.legend(['y1','y2'])
#显示所画的图
plt.show()
```

如果觉得官网看起来麻烦，可以直接百度matplotlib + 你要画的图，例如matplotlib散点图
