```python
# 第三方包导入
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# 导入数据集
data = pd.read_csv("./data/Advertising.csv", index_col=0)
# 查看部分数据
data.head()
```

```python
# 可视化数据，特征与目标值之间的联系
sns.pairplot(data, x_vars=['TV','Radio','Newspaper'], y_vars='Sales', size=7, aspect=0.8)
```

```python
# 多了一个参数，kind=‘reg’,添加后会多一条拟合曲线和95%置信带
sns.pairplot(data, x_vars=['TV','Radio','Newspaper'], y_vars='Sales', size=7, aspect=0.8, kind='reg')
```

```python
feature_cols = ['TV', 'Radio', 'Newspaper']

# 取feature数据，将特征和y分离
X = data[feature_cols]
# 打印特征向量的shape,第一个值表示有多少数据样本，第二个值是指每个样本有多少的特征，这里即为3
print(X.shape)
X.head()
```

```python
# 把目标值y单独拿出来
y = data['Sales']
# (200，)即等于（200，1）
print(y.shape)
y.head()
```

```python
# 加载sklearn里的包，用于将原始样本划分训练集特征，训练集y，测试集特征，测试集y
from sklearn.model_selection import train_test_split
# random_state表示一个随机状态，便于还原
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 从sklearn里的线性模型里加载线性回归模型
from sklearn.linear_model import LinearRegression
# 初始化一个模型
linreg = LinearRegression()
# 使用训练数据来训练该模型
linreg.fit(X_train, y_train)
```

# 模型参数以及方法更多查看见ref，写的很详细
https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
# 截距
print(linreg.intercept_)
# 系数（这里有三个特征，对应系数的维度也为3）
print(linreg.coef_)
```

```python
# 使用训练好的模型来预测
y_pred = linreg.predict(X_test)
```

```python
y_pred
```

# 回归模型的常用评价指标 

```python
from sklearn import metrics
import numpy as np
# MAE 平方绝对误差
print("MAE:",metrics.mean_absolute_error(y_test, y_pred))
```

```python
# MSE 均方误差
print("MSE:",metrics.mean_squared_error(y_test, y_pred))
```

```python
# RMSE 均方根误差
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
